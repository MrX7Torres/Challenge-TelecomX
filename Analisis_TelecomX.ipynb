##### Grafica de Densidad (En base a Cantidad) "Tipo Histograma"

```python
fig, axs = plt.subplots(filas, columnas, figsize=(15, 6 * filas))
fig.subplots_adjust(hspace=0.4, wspace=0.3)
fig.suptitle('Cantidad de Usuarios por Cargos y Tiempo', fontsize=20)

axs = axs.ravel()

for i, col in enumerate(columnas_numericas):
    sns.histplot(data=df_analisis, x=col, hue='Usuario Activo',
                 ax=axs[i],
                 palette=['green', 'red'],
                 element="poly",
                 fill=True,
                 stat="count",
                 bins=20,
                 alpha=0.3)

    axs[i].set_title(f'Distribución de {col}', fontsize=14)
    axs[i].set_ylabel('Cantidad de Usuarios')
    axs[i].set_xlabel(col)
    axs[i].grid(True, alpha=0.3)

for i in range(num_graficas, len(axs)):
    axs[i].axis('off')

plt.show()
```

### Arreglo de variables (Para convertirlas a Numericas) a partir de Mapas Logicos

```python
mapa_logico = {
    'No': 0,
    'Si': 1
}
mapa_contrato = {
    'Month-to-month': 0,
    'One year': 1,
    'Two year': 2
}

df_final = df_corr.select_dtypes(include=['number'])

for col in df_corr.columns:
    if df_corr[col].dtype == 'object':
        df_corr[col] = df_corr[col].replace(mapa_logico)
        if 'Contrato' == col:
          df_corr['Contrato'] = df_corr['Contrato'].map(mapa_contrato)


df_numerico = df_corr.select_dtypes(include=['number'])

print(df_numerico['Socio'].value_counts())
```

### Mapas de Calor

#### Correlacion de todas las Variables Numericas

```python
plt.figure(figsize=(12, 10))

sns.heatmap(df_numerico.corr(),
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            linewidths=0.5)

plt.title("Matriz de Correlación de Variables", fontsize=16)
plt.show()
```

#### Correlacion de las Variables con los Usuarios Activos

```python
plt.figure(figsize=(8, 10))
sns.heatmap(df_numerico.corr()[['Usuario Activo']].sort_values(by='Usuario Activo', ascending=False),
            annot=True,
            cmap='coolwarm',
            vmin=-1, vmax=1)

plt.title("Correlación de Variables con la BAJA (Usuario Activo)")
plt.show()
```
