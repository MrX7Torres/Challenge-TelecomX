## Carga y Manipulacion de JSON

```python
import pandas as pd
```

```python
datos = '/content/drive/Othercomputers/My Computer/SCHOOL/X Curso ONE/CHALLENGES/Challenge-TelecomX/TelecomX_Data.json'
```

```python
datos
```

```python
df = pd.read_json(datos)
df
```

```python
df_customer = pd.json_normalize(df['customer'].to_list())
df_customer
```

```python
df_phone = pd.json_normalize(df['phone'].to_list())
df_phone
```

```python
df_internet = pd.json_normalize(df['internet'].to_list())
df_internet
```

```python
df_account = pd.json_normalize(df['account'].to_list())
df_account
```

Codigo para mostrar todas las columnas

```python
pd.set_option('display.max_columns', None)
```

```python
df_final = pd.concat([df[['customerID', 'Churn']], df_customer, df_phone, df_internet, df_account], axis=1)
df_final
```
