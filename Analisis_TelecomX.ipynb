### Recuento de evasion por variables categoricas

```python
evasion_genero = df['Usuario_Activo'].groupby(df['genero']).value_counts()
```

```python
evasion_genero = evasion_genero.to_frame()
```

```python
evasion_genero = evasion_genero.rename(columns={'count':'Cantidad'})
```

```python
evasion_genero = evasion_genero.rename(index={0:'De Baja', 1:'Activo'})
```

```python
evasion_genero = evasion_genero.reset_index()
evasion_genero = evasion_genero.set_index('genero')
```

```python
evasion_genero = evasion_genero.rename(index={'Male':'Hombre', 'Female':'Mujer'})
```

```python
evasion_genero
```

```python
evasion_genero.plot(kind='bar', figsize=(10, 6), grid=True, title='Usuarios Activos vs De Baja (Por Genero)');
```

```python
evasion_genero.loc['Hombre', ['Usuario_Activo', 'Cantidad']]
```

```python
generos = ['Hombre', 'Mujer']
etiquetas_ejeX = ['Activos', 'De Baja']

fig, axs = plt.subplots(1, 2, figsize=(15, 8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Usuarios Activos vs De Baja (Por Genero)', fontsize=16)

for i, ax in enumerate(axs.flat):
    valores = evasion_genero.loc[generos[i], 'Cantidad'].values
    ax.bar(etiquetas_ejeX, valores, color=['green', 'red'])

    ax.set_title(f'{generos[i]}', fontsize=14)
    ax.set_ylabel('Cantidad de Usuarios')
    ax.grid(axis='y', alpha=0.5)

ymin = 0
ymax = 3000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

```python
import seaborn as sns
import math
```

```python
df['Usuario_Activo'] = df['Usuario_Activo'].replace({0:'De Baja', 1:'Activo'})
```

```python
df
```

```python
columnas_analisis = ['genero', 'Ciudadano_Mayor', 'Tipo_Servicio_Internet', 'Contrato', 'Metodo_Pago']
num_graficas = len(columnas_analisis)
columnas = 2
filas = math.ceil(num_graficas / columnas)

fig, axs = plt.subplots(filas, columnas, figsize=(15, 5 * filas))
fig.subplots_adjust(hspace=0.4, wspace=0.3)
fig.suptitle("Analisis por de Usuarios Activos por Variable Categorica", fontsize=18)

axs = axs.ravel()

for i, col in enumerate(columnas_analisis):
  sns.countplot(x= col, hue='Usuario_Activo', data=df, ax=axs[i], palette=['green', 'red'])

  axs[i].set_title(f'Distribución por {col}', fontsize=14)
  axs[i].set_ylabel('Cantidad de Clientes')
  axs[i].set_xlabel('')
  axs[i].legend(title='Estado', labels=['Activo', 'Baja'])

for i in range(num_graficas, len(axs)):
    axs[i].axis('off')

plt.show()
```
